```python
import pandas as pd
pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 200)
import matplotlib.pyplot as plt
import datetime as dt
```

```python
from ebaysdk.finding import Connection as finding
from bs4 import BeautifulSoup
from PIL import Image
import requests
from io import BytesIO
api_key='ForestPo-forestco-PRD-2c970d9ce-4fbe6074' 

def scraper(api_key,page_number, start=0, save_img=False):
    print("scraper start :", start)
    #keywords = input('what keywords do you want to search?')
    keywords = 'men shirt'
    api = finding(appid=api_key,config_file=None)
    api_request = {
        'itemFilter':[{
            'name': 'HideDuplicateItems',
            'value': True
        }],
        'keywords':keywords,
        'outputSelector':'GalleryInfo',
        'paginationInput': {
             'entriesPerPage':100,
             'pageNumber': page_number
         },
        #'sortOrder': 'StartTimeNewest'
    }

    response = api.execute('findItemsAdvanced',api_request)
    products = response.dict()
    #print(products)

    soup = BeautifulSoup(response.content,'lxml')

    totalentries = int(soup.find('totalentries').text)
    print(totalentries)
    items = soup.find_all('item')
    
    picture_urls = [item.find('galleryurl').text for item in items if(item.find('galleryurl'))]
    start_dates = [item.find('starttime').text for item in items if(item.find('galleryurl'))]
    c = start
    if save_img:
        for pic in picture_urls:
            response = requests.get(pic)
            img = Image.open(BytesIO(response.content))
            save_path = 'img/'+str(c)+'.png'
            img.save(save_path)
            print('saving image',c)
            c+=1
    df = pd.DataFrame({
        'date': start_dates,
        'c': range(start, c)
    })
    print("scrapper end: ", c)
    return df, c
```

```python
data_list = []
start = 0
for i in range(100):
    print(i)
    data, start = scraper(api_key,1, start=start, save_img=True)
    data_list.append(data)
```

```python
data = pd.concat(data_list).reset_index(drop=True).rename(columns={'c': 'img'})
```

```python
data.date = pd.to_datetime(data.date)
```
