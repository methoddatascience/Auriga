```python
from ebaysdk.finding import Connection as finding
from bs4 import BeautifulSoup
from PIL import Image
import requests
from io import BytesIO
api_key= 

def scraper(api_key,page_number):
    keywords = input('what keywords do you want to search?')
    api = finding(appid=api_key,config_file=None)
    api_request = {'keywords':keywords,
                   'outputSelector':'GalleryInfo',
                   'paginationInput': {
                       'entriesPerPage':'100',
                       'pageNumber': page_number
                   }
                  }

    response = api.execute('findCompletedItems',api_request)

    soup = BeautifulSoup(response.content,'lxml')

    totalentries = int(soup.find('totalentries').text)
    items = soup.find_all('item')
    items = soup.find_all('item')
    picture_urls = [item.find('galleryurl').text for item in items]
    start_dates = [item.find('starttime').text for item in items]
    for idx,pic in enumerate(picture_urls):
        response = requests.get(pic)
        img = Image.open(BytesIO(response.content))
        save_path = './images/'+ start_dates[idx] +'.jpg'
        img.save(save_path)
        print('saving image',idx)
    return start_dates




```

```python
dates = scraper(api_key,30)
```
